```python
df = pd.read_csv('data/wfp_food_prices_ind.csv', low_memory=False)
df.columns
```

```python
df = df[['date', 'commodity', 'price']]
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# Remove rows where price is not numeric
df = df[pd.to_numeric(df['price'], errors='coerce').notna()]
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
df['price'] = df['price'].astype(float)
df['date'] = pd.to_datetime(df['date'], format='mixed')
```

```python
df.info()
```

```python
df.describe()
```

```python
df['commodity'].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

rice_df = df[df['commodity'] == 'Rice']

rice_monthly = rice_df.groupby(rice_df['date'].dt.year)['price'].mean()

plt.figure()
plt.plot(rice_monthly.index, rice_monthly.values)
plt.xlabel('Year')
plt.ylabel('Average Price (INR)')
plt.title('Average Rice Price Trend in India')
plt.show()
```

```python
top_commodities = ['Rice', 'Wheat', 'Sugar']

filtered = df[df['commodity'].isin(top_commodities)]

trend = filtered.groupby(
    [filtered['date'].dt.year, 'commodity']
)['price'].mean().unstack()

trend.plot()
plt.xlabel('Year')
plt.ylabel('Average Price (INR)')
plt.title('Food Price Trends of Major Commodities in India')
plt.show()
```

```python
df = df[df['date'].dt.year <= 2019]
```

```python
import matplotlib.pyplot as plt

rice_df = df[df['commodity'] == 'Rice']

rice_monthly = rice_df.groupby(rice_df['date'].dt.year)['price'].mean()

plt.figure()
plt.plot(rice_monthly.index, rice_monthly.values)
plt.xlabel('Year')
plt.ylabel('Average Price (INR)')
plt.title('Average Rice Price Trend in India')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
df[df['commodity'].isin(['Rice', 'Wheat', 'Sugar'])].boxplot(
    column='price',
    by='commodity'
)
plt.title('Price Distribution of Major Food Commodities')
plt.suptitle('')
plt.xlabel('Commodity')
plt.ylabel('Price (INR)')
plt.show()
```
